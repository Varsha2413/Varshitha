```python
# !pip install matplotlib
import matplotlib.pyplot as pit
import numpy as np
day=[1,2,3,4,5]
date=['jun23','jun24','jun25','jul1','jul2']
temp=[28,30,33,35,25]
wind=[22,24,25,23,26]
ypos=np.arange(len(day))
pit.bar(ypos+0.2,temp,color='y',width=0.4)
pit.bar(ypos-0.2,wind,color='g',width=0.4)
pit.xticks(ypos,day)
pit.xlabel("days in week")
pit.ylabel("temperature")
pit.title("days and temperature")
```
